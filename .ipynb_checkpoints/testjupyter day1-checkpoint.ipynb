```python
from qiskit import QuantumCircuit 
from qiskit_aer import AerSimulator 
from qiskit.visualization import plot_histogram
```

```python
#porta inicial sem nenhuma porta - estado |0>

qc = QuantumCircuit(1,1)
qc.measure(0,0)

simulator = AerSimulator()
resultado = simulator.run(qc, shots=1000).result()
contagens = resultado.get_counts()

print("Resultados:", contagens)
plot_histogram(contagens)
```

```python
qc2 = QuantumCircuit(1,1)
qc2.h(0)
qc2.measure(0,0)

resultado2 = simulator.run(qc2, shots=1000).result()
contagens2 = resultado2.get_counts()

print("Resultados:" , contagens2)
plot_histogram(contagens2)
```
