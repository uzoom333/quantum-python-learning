```python
import numpy as np
from numpy.linalg import matrix_rank
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# Exercicio 1: v1,v2,v3 independentes? v1,v2,v3,v4 dependentes?
v1 = np.array([1, 0, 0])
v2 = np.array([1, 1, 0])
v3 = np.array([1, 1, 1])
v4 = np.array([2, 3, 4])

# Teste v1, v2, v3
A3 = np.column_stack([v1, v2, v3])
print('=== v1, v2, v3 ===')
print(f'Matriz:\n{A3}')
print(f'Rank: {matrix_rank(A3)} -> {"Independentes" if matrix_rank(A3) == 3 else "Dependentes"}')

# Teste v1, v2, v3, v4
A4 = np.column_stack([v1, v2, v3, v4])
print('\n=== v1, v2, v3, v4 ===')
print(f'Matriz:\n{A4}')
print(f'Rank: {matrix_rank(A4)} -> {"Independentes" if matrix_rank(A4) == 4 else "Dependentes"}')

# v4 como combinacao de v1, v2, v3
coefs = np.linalg.solve(A3, v4)
print(f'\nv4 = {coefs[0]:.0f}*v1 + {coefs[1]:.0f}*v2 + {coefs[2]:.0f}*v3')
```

```python
# Exercicio 2: Maior numero de vetores independentes
v1 = np.array([1, -1, 0, 0])
v2 = np.array([1, 0, -1, 0])
v3 = np.array([1, 0, 0, -1])
v4 = np.array([0, 1, -1, 0])
v5 = np.array([0, 1, 0, -1])
v6 = np.array([0, 0, 1, -1])

A = np.column_stack([v1, v2, v3, v4, v5, v6])
print(f'Matriz 4x6:\n{A}')
print(f'\nRank: {matrix_rank(A)}')
print(f'Maior numero de vetores independentes: {matrix_rank(A)}')
print(f'Dimensao do espaco gerado: {matrix_rank(A)}')

# Verificar dependencias
print(f'\nv4 = v2 - v1? {np.allclose(v4, v2 - v1)}')
print(f'v5 = v3 - v1? {np.allclose(v5, v3 - v1)}')
print(f'v6 = v3 - v2? {np.allclose(v6, v3 - v2)}')
```

```python
# Exercicio 3: Provar que se a=0, d=0 ou f=0, colunas de U sao dependentes

# Caso 1: a = 0
U1 = np.array([[0, 2, 3], [0, 1, 5], [0, 0, 2]])
print(f'Caso 1 (a=0):\n{U1}')
print(f'Rank: {matrix_rank(U1)} < 3 -> Dependentes\n')

# Caso 2: d = 0
U2 = np.array([[3, 2, 3], [0, 0, 5], [0, 0, 2]])
print(f'Caso 2 (d=0):\n{U2}')
print(f'Rank: {matrix_rank(U2)} < 3 -> Dependentes\n')

# Caso 3: f = 0
U3 = np.array([[3, 2, 3], [0, 1, 5], [0, 0, 0]])
print(f'Caso 3 (f=0):\n{U3}')
print(f'Rank: {matrix_rank(U3)} < 3 -> Dependentes\n')

# Todos pivos nao-zero: independentes
U_ok = np.array([[3, 2, 3], [0, 1, 5], [0, 0, 2]])
print(f'Todos pivos != 0:\n{U_ok}')
print(f'Rank: {matrix_rank(U_ok)} = 3 -> Independentes')
```

```python
# Conexao com Quantum: Base computacional de 2 qubits
# 2 qubits vivem em C^4 (dimensao 4)
# Base: |00>, |01>, |10>, |11>

ket_00 = np.array([1, 0, 0, 0])
ket_01 = np.array([0, 1, 0, 0])
ket_10 = np.array([0, 0, 1, 0])
ket_11 = np.array([0, 0, 0, 1])

base_2qubits = np.column_stack([ket_00, ket_01, ket_10, ket_11])
print('Base computacional de 2 qubits:')
print(f'Matriz:\n{base_2qubits}')
print(f'Rank: {matrix_rank(base_2qubits)} -> dimensao = 4')
print(f'\nE a matriz identidade! Cada qubit state e um vetor da base.')

# Bell State: combinacao da base
bell = (ket_00 + ket_11) / np.sqrt(2)
print(f'\nBell State |Phi+> = (|00> + |11>) / sqrt(2)')
print(f'Vetor: {np.round(bell, 4)}')
print(f'Probabilidades: |00>={bell[0]**2:.1%}, |01>={bell[1]**2:.1%}, |10>={bell[2]**2:.1%}, |11>={bell[3]**2:.1%}')
```

```python
# Criar Bell State no Qiskit e verificar
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0,1], [0,1])

simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts()
print(f'Qiskit Bell State resultado: {counts}')
print('50% |00> e 50% |11> - qubits emaranhados!')
plot_histogram(counts)
```
